```python
import pandas as pd

# Read the updated dataset
data = pd.read_csv('Updated Employee Dataset.csv')

# Filter for inexpert employees who need mentors
inexpert_data = data[data['Employee Type'] == 'Inexpert']

# Define a function to find an expert mentor for an inexpert employee
def assign_mentor(row):
    mentors = data[(data['Employee Type'] == 'Expert') & (data['Skill Set'] == row['Skill Set']) & (data['SuccessFull Completions Percentage'] >= 40)]
    
    if not mentors.empty:
        mentor = mentors.iloc[0]
        return pd.Series({'Mentor ID': mentor['Emp.id'], 'Mentor Name': mentor['Employee Name']})
    else:
        return pd.Series({'Mentor ID': None, 'Mentor Name': None})

# Apply the mentor assignment function to each inexpert employee
mentors_df = inexpert_data.apply(assign_mentor, axis=1)

# Combine the mentor information with the inexpert employee data
inexpert_data = pd.concat([inexpert_data, mentors_df], axis=1)

# Save the updated dataset with mentor assignments
inexpert_data.to_csv('Inexpert Employee Dataset with Mentors.csv', index=False)

print("Mentor assignment completed and saved to 'Inexpert Employee Dataset with Mentors.csv'.")
```

```python
import pandas as pd

# Read the Employee Dataset.csv file
employee_data = pd.read_csv("Employee dataset.csv")

# Define a mapping function
def map_employee_to_category(row):
    experience = row['Experience']
    tasks_completed = row['No of Tasks Completed']

    if experience < 4:
        return 'Inexpert'
    elif experience > 4 and tasks_completed > 50:
        return 'Expert'
    else:
        return 'Undefined'

# Apply the mapping function to create a new 'Category' column
employee_data['Category'] = employee_data.apply(map_employee_to_category, axis=1)

# Filter out employees with 'Undefined' category
employee_data = employee_data[employee_data['Category'] != 'Undefined']

# Separate experts and inexpert employees
experts = employee_data[employee_data['Category'] == 'Expert']
inexpert = employee_data[employee_data['Category'] == 'Inexpert']

# Assign inexpert employees to experts
experts['Inexpert_Assigned'] = experts.index // 3  # Assign  inexpert employees to each expert

# Print the resulting DataFrame
print("Expert Employees:")
print(experts)
print("\nInexpert Employees:")
print(inexpert)
```

```python
import pandas as pd

# Read the Employee and Tasks datasets
employee_data = pd.read_csv("Employee dataset.csv")
tasks_data = pd.read_csv("Tasks Dataset.csv")

# Define the mapping function to categorize employees
def map_employee_to_category(row):
    experience = row['Experience']
    tasks_completed = row['No of Tasks Completed']

    if experience < 3:
        return 'Inexpert'
    elif experience > 3 and tasks_completed > 40:
        return 'Expert'
    else:
        return 'Undefined'

# Apply the mapping function to create a new 'Category' column
employee_data['Category'] = employee_data.apply(map_employee_to_category, axis=1)

# Calculate skills similarity
employee_data['Skills Set'] = employee_data['Skill Set'].str.split(',')
tasks_data['Task Requirement Skills'] = tasks_data['Task Requirement Skills'].str.split(',')

employee_data['Skills Similarity'] = employee_data.apply(
    lambda row: sum(1 for skill in row['Skills Set'] if skill in tasks_data.iloc[0]['Task Requirement Skills']), axis=1
)

# Filter tasks for experts
experts = employee_data[employee_data['Category'] == 'Expert']

task_assignment = {}
for expert_index, expert_row in experts.iterrows():
    expert_name = expert_row['Employee Name']
    expert_skills_similarity = expert_row['Skills Similarity']
    
    # Filter tasks based on skills similarity
    eligible_tasks = tasks_data[tasks_data['Task Requirement Skills'].apply(lambda x: expert_skills_similarity / len(x)) > 0.3]
    
    # Sort tasks by level of urgency and due date
    eligible_tasks = eligible_tasks.sort_values(by=['Level of Urgency', 'Due Date in days'])
    
    assigned_tasks = []
    avg_task_time = expert_row['Avg Time To Finish Tasks (in Days)']
    for _, task_row in eligible_tasks.iterrows():
        task_id = task_row['Task id']
        if task_id not in task_assignment and len(assigned_tasks) <1:
            assigned_tasks.append(task_id)
            task_assignment[task_id] = expert_name
            # Adjust the expert's average task time with the new task time
            expert_row['Avg Time To Finish Tasks (in Days)'] = (avg_task_time * len(assigned_tasks) + task_row['Due Date in days']) / (len(assigned_tasks) + 1)

            print(f"Task ID: {task_id} -> Assigned to Expert: {expert_name}")
# Print the task assignments
for task_id, expert_name in task_assignment.items():
    print(f"Task ID: {task_id} -> Assigned to Expert: {expert_name}")
```
